```python
import pandas as pd
import matplotlib.pyplot as plt

# Punjab Gas Data
data = {
    "Year": [1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007],
    "Total": [657951, 803125, 810065, 844640, 896615, 1200148, 1427550, 1506825, 1505163, 1560922],
    "Household": [203390, 256467, 222559, 227132, 241393, 246742, 277593, 272147, 305514, 334472],
    "Commercial": [37423, 43022, 34561, 38250, 39441, 42539, 48875, 52653, 56996, 59388],
    "Industrial": [413201, 501600, 551230, 577857, 613697, 908489, 1098383, 1179609, 1141575, 1164719],
    "Others": [3937, 2036, 1715, 1401, 2084, 2378, 2699, 2416, 1078, 2343]
}

punjab_gas_df = pd.DataFrame(data)

# Step 1: Descriptive Statistics
print("Descriptive Statistics:")
print(punjab_gas_df.describe())

# Step 2: Line Graph of Total Consumption Over Time
plt.figure(figsize=(10, 6))
plt.plot(punjab_gas_df["Year"], punjab_gas_df["Total"], marker='o', label="Total Consumption", color='blue')
plt.title("Total Gas Consumption in Punjab Over Years")
plt.xlabel("Year")
plt.ylabel("Consumption (mm cf)")
plt.grid(True)
plt.legend()
plt.show()

# Step 3: Bar Chart of Sector-Wise Consumption
punjab_gas_df.set_index("Year")[["Household", "Commercial", "Industrial", "Others"]].plot(
    kind="bar", stacked=True, figsize=(12, 8))
plt.title("Sector-Wise Gas Consumption in Punjab Over Years")
plt.xlabel("Year")
plt.ylabel("Consumption (mm cf)")
plt.legend(title="Sector")
plt.show()
```
